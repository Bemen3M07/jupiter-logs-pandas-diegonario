```python
import logging

# Configuració dels Handlers
info_handler = logging.FileHandler('info.log', mode='w')  # Sobreescriure o crear fitxer info.log
error_handler = logging.FileHandler('error.log', mode='w')  # Sobreescriure o crear fitxer error.log
csv_handler = logging.FileHandler('logs.csv', mode='w')  # Sobreescriure o crear fitxer logs.csv
console_handler = logging.StreamHandler()  # Handler per mostrar a pantalla

# Configuració dels Formatters
info_format = logging.Formatter('%(asctime)s - %(levelname)s - %(message)s')
error_format = logging.Formatter('%(asctime)s - %(levelname)s - %(message)s')
csv_format = logging.Formatter('%(asctime)s,%(levelname)s,%(message)s')
console_format = logging.Formatter('%(levelname)s: %(message)s')

info_handler.setFormatter(info_format)
error_handler.setFormatter(error_format)
csv_handler.setFormatter(csv_format)
console_handler.setFormatter(console_format)

# Configuració del Logger principal
main_logger = logging.getLogger('main_logger')
main_logger.setLevel(logging.DEBUG)

main_logger.addHandler(info_handler)  # Missatges INFO al fitxer info.log
main_logger.addHandler(error_handler)  # Missatges ERROR al fitxer error.log
main_logger.addHandler(console_handler)  # Missatges a la consola

# Configuració d'un Logger especial per a CSV
csv_logger = logging.getLogger('csv_logger')
csv_logger.setLevel(logging.INFO)
csv_logger.addHandler(csv_handler)  # Missatges INFO al fitxer logs.csv

# Generar 20 registres de logs
for i in range(20):
    main_logger.info(f'Registre INFO número {i + 1}')
    main_logger.error(f'Registre ERROR número {i + 1}')
    csv_logger.info(f'CSV INFO número {i + 1}')
```
