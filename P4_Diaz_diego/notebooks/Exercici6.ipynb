```python

import pandas as pd
import matplotlib.pyplot as plt

# Carregar el fitxer CSV
dades = pd.read_csv("dades_alumnes.csv")

# Afegir una columna amb la mitjana de cada alumne
dades["Mitjana_Alumne"] = dades[["M01", "M04", "M03", "M05"]].mean(axis=1)

# Gràfica 1: Distribució de les mitjanes per alumne
plt.figure(figsize=(10, 6))
plt.bar(dades["Alumne"], dades["Mitjana_Alumne"], color='skyblue')
plt.title("Distribució de les mitjanes per alumne")
plt.xlabel("Alumne")
plt.ylabel("Mitjana")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.savefig("grafic1_distribucio.png")
plt.show()

# Gràfica 2: Percentatges d'aprovats i suspesos
aprovats = (dades["Mitjana_Alumne"] >= 5).sum()
suspesos = (dades["Mitjana_Alumne"] < 5).sum()

plt.figure(figsize=(6, 6))
plt.pie([aprovats, suspesos], labels=["Aprovats", "Suspesos"], autopct="%1.1f%%", colors=['green', 'red'])
plt.title("Percentatges d'aprovats i suspesos")
plt.savefig("grafic2_percentatges.png")
plt.show()

# Gràfica 3: Evolució de les notes per alumne
plt.figure(figsize=(10, 6))
for alumne in dades["Alumne"]:
    plt.plot(["M01", "M04", "M03", "M05"], dades.loc[dades["Alumne"] == alumne, ["M01", "M04", "M03", "M05"]].values[0], label=alumne)
plt.title("Evolució de les notes per alumne")
plt.xlabel("Mòduls")
plt.ylabel("Nota")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.savefig("grafic3_evolucio.png")
plt.show()

# Gràfica 4: Nota més alta i més baixa
nota_mes_alta = dades["Mitjana_Alumne"].max()
nota_mes_baixa = dades["Mitjana_Alumne"].min()

plt.figure(figsize=(6, 4))
plt.bar(["Nota més alta", "Nota més baixa"], [nota_mes_alta, nota_mes_baixa], color=["gold", "blue"])
plt.title("Comparació de la nota més alta i més baixa")
plt.ylabel("Nota")
plt.savefig("grafic4_comparacio.png")
plt.show()
```
